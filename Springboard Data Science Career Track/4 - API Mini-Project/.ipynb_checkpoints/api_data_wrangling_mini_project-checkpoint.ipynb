These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
#call Quandl API for 2017, request return as json, convert to dict
url2017 = 'https://www.quandl.com/api/v3/datasets/FSE/AFX_X.json?start_date=2017-01-01&end_date=2017-12-31&api_key=' + API_KEY
r2017 = requests.get(url2017)
json_data_2017 = r2017.json()
json_2017_dict = dict(json_data_2017)
print(type(json_2017_dict))
```

```python
#evaluate dict

print(type(json_2017_dict['dataset']))
print(json_2017_dict.keys())
print(len(json_2017_dict['dataset']['column_names']))
for key in json_2017_dict['dataset'].keys():
    print(str(key) + ": " + str(type(json_2017_dict['dataset'][key])))
```

```python
#A look at the number of dates in the dict 

dates_2017 = []
for index in json_2017_dict['dataset']['data']:
    dates_2017.append(index[0])
len(dates_2017)
```

255 is in line with the number of dates the FSX was open in 2017 when you factor in weekends and holidays

```python
#Determine Max Opening prices and Min Opening prices for the years
#created a list of the opening prices and applied max, min while accounting for None values

open_values_2017 = []
for index in json_2017_dict['dataset']['data']:
    open_values_2017.append(index[1])
print('The maximum opening price was: ' + str((max(filter(None,open_values_2017)))))
print('The minimum opening price was: ' + str((min(filter(None,open_values_2017)))))
```

```python
#Determine the largest change in a day based on high and low prices

high_values_2017 = []
low_values_2017 = []
high_low_diff_2017_abs = []
for index in json_2017_dict['dataset']['data']:
    high_values_2017.append(index[2])
    low_values_2017.append(index[3])
for high,low in zip(high_values_2017,low_values_2017):
    high_low_diff_2017_abs.append(abs(high-low))
print('The largest change in a day was: ' + str(round(max(filter(None,high_low_diff_2017_abs)),2)))
```

```python
#Determine the largest change between days based closing price

closing_values_2017 = []
closing_diff_2017_abs = []
for index in json_2017_dict['dataset']['data']:
    closing_values_2017.append(index[4])
for next_day, this_day in zip(high_values_2017[1:],low_values_2017[:-1]):
    closing_diff_2017_abs.append(abs(next_day - this_day))
print('The largest change in closing from one day to the next was: ' + str(round(max(filter(None,closing_diff_2017_abs)),2)))
```

```python
#Determine average trading volume for the year

TradingVolume_values_2017 = []
for index in json_2017_dict['dataset']['data']:
    TradingVolume_values_2017.append(index[6])
print('The average trading for 2017 was: ' + str(round(sum(TradingVolume_values_2017)/len(TradingVolume_values_2017),2)))
```

```python
#Create a function to determine the median of lists and pass Trading Volume to the function to determine the median

def finding_median(unsorted_list):
    sorted_list = sorted(unsorted_list)
    if len(sorted_list)%2 != 0:
        return sorted_list[int(len(sorted_list)/2)]
    else:
        return round((sorted_list[len(sorted_list)/2] + sorted_list[len(sorted_list)/2 + 1])/2,2)
       
finding_median(TradingVolume_values_2017)
```

Verification of results with some plots

```python
#Sanity checks of results using matplotlib and some charts
import matplotlib.pyplot as plt

#Max and Min Values
plt.boxplot(sorted(filter(None,open_values_2017)))
plt.title('AFX_X Closing Vlaues Boxplot - 2017')
plt.ylabel('AFX_X Closing Values - 2017')
plt.show()

#Largest change in a day based on high-low values
plt.plot(high_low_diff_2017_abs)
plt.xlabel('High-Low Difference in Price per day')
plt.ylabel('Day Number')
plt.title('AFX_X High-Low Difference in Price per day - 2017')
plt.show()

#Largest change between days based on closing price
plt.plot(closing_diff_2017_abs)
plt.xlabel('Different in closing prices from day to day')
plt.ylabel('Day Number')
plt.title('AFX_X Difference in closing prices from day to day - 2017')
plt.show()

#histogram to show distribution of trading volumen in 2017
plt.hist(TradingVolume_values_2017,bins=60)
plt.title('Trading Volume Distribution')
plt.show()
```
